```python
# Import Dependencies
import pandas as pd
```

```python
# Create a reference the CSV file desired
csv_path = "Resources/datasubset.csv"

# Read the CSV into a Pandas DataFrame
Original_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
Original_df.head()
```

```python
Original_df.describe()
```

```python
Type_df = Original_df[["UNITID", "Institution_Name", "Public_Privatenp_Privatefp"]]
Type_df.loc[Type_df.Public_Privatenp_Privatefp == 1, 'Type'] = 'Public'  
Type_df.loc[Type_df.Public_Privatenp_Privatefp == 2, 'Type'] = 'Private Non-Profit'  
Type_df.loc[Type_df.Public_Privatenp_Privatefp == 3, 'Type'] = 'Private For-Profit' 
Type_df.head()
```

```python
Type_df.to_csv("Output/Type.csv", index=False, header=True)
```
